# Feature Engineering for Sales Forecasting

```python
# Loading the dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

master_sales = pd.read_csv("../data/master_sales_v2.csv", parse_dates=["sale_date"])
```

```python
master_sales.head()
```

```python
# sorting the data
master_sales = master_sales.sort_values(by=['store_id', 'dept_id', 'sale_date'])
```

```python
# Time based features
master_sales['year'] = master_sales['sale_date'].dt.year
master_sales['month'] = master_sales['sale_date'].dt.month
master_sales['week'] = master_sales['sale_date'].dt.isocalendar().week
master_sales['day_of_week'] = master_sales['sale_date'].dt.dayofweek
```
